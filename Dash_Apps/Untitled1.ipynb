```python
app.layout=html.Div([dcc.Location(id='url'),
                     sidebar,
                     html.Div(id="page-content", style=CONTENT_STYLE),
])


@app.callback(Output("page-content", "children"),
              [Input("url", "pathname")])
def render_page_content(pathname):
    if pathname == "/":
        return home
    elif pathname == "/DeviceList" :
        return html.P("This is the content of DeviceList ")
    elif pathname == "/JSP/DBEM" :
        return html.P("This is the content of JSP/DBEM")
    elif pathname == "/SmartMeter" :
        return html.P("This is the content of SmartMeter")
    elif pathname == "/HSM" :
        return html.P("This is the content of HSM")
    elif pathname == "/Solar" :
        return html.P("This is the content of Solar")
    elif pathname == "/DBEM+JSP" :
        return html.P("This is the content of DBEM+JSP")
    elif pathname == "/HeatMap" :
        return html.P("This is the content of HeatMap")
    elif pathname == "/Reports" :
        return html.P("This is the content of Reports")
    elif pathname == "/DataHours":
        return html.P("This is the content of Data Hours")
    elif pathname == "/MetaData" :
        return html.P("This is the content of MetaData")
    elif pathname == "/Devices-Info" :
        return html.P("This is the content of Devices-Info")

    # If the user tries to reach a different page, return a 404 message
    return html.Div(
        [
            html.H1("404: Not found", className="text-danger"),
            html.Hr(),
            html.P(f"The pathname {pathname} was not recognised..."),
        ],
        className="p-3 bg-light rounded-3",
    )

@app.callback(Output("home-page",'children'),
             [Input('submit1','n_clicks')],
             [State('dropdown3','value')],
             [State('dropdown4','value')])
def home_out(n_clicks,val1,val2):
    return "Value1 is {}, Value2 is {}, No. of clicks are {}".format(val1,val2,n_clicks)


if __name__ == "__main__":
    app.run_server(debug=True)
```
